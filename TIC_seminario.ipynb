```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear DataFrame con los datos proporcionados manualmente
data = {
    "NOMBRE": [
        "Hugo Arango", "Rosmery Arango", "Luz Karime Arango", "Paola Barbosa", "Adriana Patricia Caceres",
        "Yolanda Cardozo", "Andrea Del Pilar Castillo", "Carlos Julio Diaz", "Franklin Diaz", "Juan Carlos Diaz",
        "Jose Joaquin Fajardo", "Maryuri Gomez", "Jhon Jawer Gomez", "Pablo Gutierrez", "Ana Milena Londoo",
        "Emilio Martin", "Olga Luca Martinez", "German Polanco", "Jenny Paola Rios", "Wilson Valbuena"
    ],
    "Nota1": [83, 100, 67, 67, 83, 83, 83, 100, 100, 100, 67, 83, 83, 83, 83, 100, 67, 67, 100, 50],
    "Nota2": [100]*20,
    "Porcentaje TIC": [1.7, 0.0, 1.7, 1.7, 1.7, 1.7, 1.7, 0.0, 0.0, 0.0, 3.3, 1.7, 1.7, 1.7, 1.7, 0.0, 3.3, 3.3, 0.0, 5.0]
}

df = pd.DataFrame(data)

# Calcular estadsticas bsicas
stats = df.describe()

# Crear grfica de distribucin de Nota1 y Nota2
plt.figure(figsize=(10, 6))
sns.histplot(df["Nota1"], bins=6, kde=True, label="Primera Evaluacin", color="skyblue")
sns.histplot(df["Nota2"], bins=6, kde=True, label="Segunda Evaluacin", color="salmon", alpha=0.6)
plt.title("Distribucin de Notas de Evaluaciones")
plt.xlabel("Nota (%)")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("/mnt/data/distribucion_notas.png")
plt.close()

# Crear grfica de mejora entre nota 1 y 2
df["Mejora"] = df["Nota2"] - df["Nota1"]

plt.figure(figsize=(10, 6))
sns.barplot(x="NOMBRE", y="Mejora", data=df, palette="viridis")
plt.xticks(rotation=90)
plt.title("Mejora entre Primera y Segunda Evaluacin")
plt.ylabel("Mejora (%)")
plt.tight_layout()
plt.savefig("/mnt/data/mejora_evaluaciones.png")
plt.close()

# Grfica de porcentaje TIC
plt.figure(figsize=(10, 6))
sns.barplot(x="NOMBRE", y="Porcentaje TIC", data=df, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Porcentaje de TIC aplicado por participante")
plt.ylabel("TIC (%)")
plt.tight_layout()
plt.savefig("/mnt/data/porcentaje_tic.png")
plt.close()

stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ruta local del archivo
ruta = r"C:\RICHARD\FDL\Sena\Prueba diagnostica.xlsx"

# Verificar si el archivo existe
if os.path.exists(ruta):
    df = pd.read_excel(ruta)
    print("Archivo cargado correctamente ")

    # 1. Histograma de notas
    plt.figure(figsize=(8, 5))
    sns.histplot(df["Nota"].dropna(), bins=7, kde=True, color="steelblue")
    plt.title("Distribucin de Notas")
    plt.xlabel("Nota")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 2. Dispersin Duracin vs Puntuacin
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x="Duracin (Minutos)", y="Puntuacin", hue="Nota", palette="Spectral")
    plt.title("Duracin vs Puntuacin")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 3. Promedio de Nota por Curso
    plt.figure(figsize=(10, 6))
    promedio_curso = df.groupby("Cdigo del curso")["Nota"].mean().sort_values(ascending=False)
    sns.barplot(x=promedio_curso.index, y=promedio_curso.values, palette="coolwarm")
    plt.title("Nota Promedio por Cdigo de Curso")
    plt.xlabel("Cdigo del curso")
    plt.ylabel("Nota promedio")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

else:
    print(" Ruta invlida o archivo no encontrado.")
```
